```python
# IMPORTACOES
import pytz
from datetime import datetime
tz = pytz.timezone('America/Sao_Paulo')

# VARIAVEIS
vendas = {}
clientes = []
produtos = []

menu = [
 'Cadastrar cliente',
 'Ver dados de um cliente específico',
 'Visualizar todos os dados dos clientes cadastrados',
 'Cadastrar produto',
 'Ver dados de um produto específico',
 'Visualizar todos os dados dos produtos cadastrados',
 'Efetuar uma venda',
 'Visualizar vendas',
 'Atualizar dados de um cliente específico',
 'Atualizar dados de um produto',
 'Sair'
]
opc = 0
a = '='
aa = '-'


class Produto:
  ultimo_id_produto = 0
  
  def __init__(self, nome, preco, descricao):
    self.id = self.generate_id()
    self.nome = nome
    self.preco = preco
    self.descricao = descricao
  @classmethod
  def generate_id(self):
      self.ultimo_id_produto += 1  
      return self.ultimo_id_produto  
  
class Cliente:

  def __init__(self, nome, email, cpf):
    self.nome = nome
    self.email = email
    self.cpf = cpf
    
class Vendas:
  ultimo_id = 0
  
  def __init__(self, data, cliente, produto):
    self.id = self.ultimo_id + 1  
    self.ultimo_id = self.id  
    self.data = data
    self.cliente = cliente
    self.produto = produto
  @classmethod
  def generate_id(self):
      self.ultimo_id += 1  
      return self.ultimo_id  




# FUNCOES
# VALIDACAO
def validarCliente(nome, email, cpf):
  erro = ''
  if nome == '' or nome is None:
    erro += 'Nome Inválido\n'
  if email == '' or email is None:
    erro += 'Email Inválido\n'
  if cpf == '' or cpf is None :
    erro += 'CPF Inválido\n'
    
  return erro;
  
def validarProduto(nome, descricao):
  erro = ''
  if nome == '' or nome is None:
    erro += 'Nome Inválido\n'
  if descricao == '' or descricao is None:
    erro += 'Descrição Inválido\n'
  return erro

# SETTERS
def setCliente():
  nome = str(input('Digite o nome do cliente: '))
  email = str(input('Digite o email do cliente: '))
  cpf = str(input('Digite o CPF do cliente: '))
  v = validarCliente(nome, email, cpf)
  if v == '' :
    cliente = Cliente(nome, email, cpf)
    clientes.append(cliente)
    print("Cliente cadastrado com Sucesso!\n")
  else:
    print(v)
    
def setProduto():
  nome = str(input('Digite o nome do produto: '))
  preco = float(input('Digite o preço do produto: '))
  descricao = str(input('Digite a descrição do produto: '))
  v = validarProduto(nome, descricao)
  if v == '' :
    produto = Produto(nome, float(preco), descricao)
    produtos.append(produto)
    print("Produto cadastrado com Sucesso!\n")
  else:
    print(v)

def setVenda():
  cpf = input("Digite o CPF do Cliente: ")
  id = int(input("Digite o ID do Produto: "))
  cliente = None
  produto = None
  for c in clientes:
    if c.cpf == cpf:
        cliente = c
        break
  for p in produtos:
    if p.id == id:
      produto = p
      break
  if cliente is None or produto is None:
    print("Não foi possível efetuar a venda pois cliente/produto não foram encontrados.\n")
  else:
    num_venda = len(vendas) + 1
    data_hora_atual = datetime.now(tz)
    data = str(data_hora_atual.strftime('%d/%m/%Y %H:%M'))
    vendas[num_venda] = (data, cliente.cpf, produto.id)
    print("Venda cadastrada com sucesso!\n")
    
def getProduto():
  traco = '-'
  if len(produtos) > 0:
    print(f'\n{traco*20}\nDADOS DO PRODUTO\n {traco*20}')
    for p in produtos:
      print(f'ID: {p.id}')
      print(f'Nome: {p.nome}')
      print(f"Preço: R$ {p.preco:.2f}")
      print(f'Descrição: {p.descricao}\n\n{traco*20}')
  else:
    print("Não encontramos o Cliente!\n")
    opcao = int(input("Deseja cadastrar? ( 0 = SIM ou 1 = NAO) \n"))
    if opcao == 0:
      setCliente()  
    
def getProdutoByID():
  id = int(input("Digite o ID do Produto: "))
  for p in produtos:
    if int(p.id) == id:
      print(f'ID: {p.id}')
      print(f"Nome: {p.nome}")
      print(f"Preço: R$ {p.preco:.2f}")
      print(f"Descrição: {p.descricao}")
      return
  print("Não encontramos o Produto!\n")
  opcao = int(input("Deseja cadastrar? ( 0 = SIM ou 1 = NAO) \n"))
  if opcao == 0:
    setProduto()

def getCliente():
  traco = '-'
  if len(clientes) > 0:
    print(f'\n{traco*20}\nDADOS DO CLIENTE\n{traco*20}')
    for c in clientes:
      print(f'Nome: {c.nome}')
      print(f"Email: {c.email}")
      print(f'CPF: {c.cpf}\n\n{traco*20}')
  else:
    print("Não encontramos o Cliente!\n")
    opcao = int(input("Deseja cadastrar? ( 0 = SIM ou 1 = NAO) \n"))
    if opcao == 0:
      setCliente()  
    
def getClienteByID():
  cpf = int(input("Digite o CPF do Cliente: "))
  for c in clientes:
    if int(c.cpf) == cpf:
      print(f"Nome: {c.nome}")
      print(f"Email: {c.email}")
      print(f"CPF: {c.cpf}")
      return
  print("Não encontramos o Cliente!\n")
  opcao = int(input("Deseja cadastrar? ( 0 = SIM ou 1 = NAO) \n"))
  if opcao == 0:
    setCliente()
      
    
def getVenda():
  traco = '-'
  if len(vendas) > 0:
    print(f'\n{traco*20}\nDADOS DA VENDA\n{traco*20}')
    for idVenda, v in vendas.items():
      data, cpf, id = v
      for c in clientes:
        if c.cpf == cpf:
          nome_cliente = str(c.nome)
          break
      for p in produtos:
        if p.id == id:
          nome_produto = str(p.nome)
          preco_produto = float(p.preco)
          break
      print(f'ID: {idVenda}')
      print(f'Data: {data}')
      print(f'Cliente: {nome_cliente}')
      print(f'Produto: {nome_produto}')
      print(f"Preço: R$ {p.preco:.2f}\n\n{traco*20}")
      return
  else:
    print("Não encontramos Nenhuma Venda!\n")
    opcao = int(input("Deseja cadastrar? ( 0 = SIM ou 1 = NAO) \n"))
    if opcao == 0:
      setVenda()  
# UPDATE
def updateCliente():
  cpf = int(input("Digite o CPF do Cliente: "))
  for c in clientes:
    if int(c.cpf) == cpf:
      c.nome = str(input("Digite o novo Nome do Cliente: "))
      c.email = str(input("Digite o novo Email do Cliente: "))
      print("Cliente foi atualizado com sucesso\n")
      return
  print("Não encontramos o Cliente!\n")
  
def updateProduto():
  id = int(input("Digite o ID do Produto: "))
  for p in produtos:
    if int(p.id) == id:
      p.nome = str(input("Digite o novo Nome do Produto: "))
      p.preco = float(input("Digite o novo Preço do Produto: "))
      p.descricao = str(input("Digite a nova Descrição do Produto: "))
      print("Produto foi atualizado com sucesso\n")
      return
  print("Não encontramos o Produto!\n")
  opcao = int(input("Deseja cadastrar? ( 0 = SIM ou 1 = NAO) \n"))
  if opcao == 0:
    setVenda()  

# PRINCIPAL(CHAMA TODAS AS FUNCOES)
def main():
  #EXECUCAO DO PROGRAMA 
  print(f'{a*20}\nSeja Bem vindo(a)\n{a*20} ')
  # MENU
  for i, j in enumerate(menu):
    print(f'[{i}] {j}')
  # OPCOES
  while True:
    opc = int(input('Digite a opção desejada: '))
    if   int(opc) == 0:
      setCliente()
    elif int(opc) == 1:
      getClienteByID()
    elif int(opc) == 2:
      getCliente()
    elif int(opc) == 3:
      setProduto()
    elif int(opc) == 4:
      getProdutoByID()
    elif int(opc) == 5:
      getProduto()
    elif int(opc) == 6:
      setVenda()
    elif int(opc) == 7:
      getVenda()
    elif int(opc) == 8:
      updateCliente()
    elif int(opc) == 9:
      updateProduto()
    elif int(opc) == 10:
      print('Até a próxima! :)')
      break;
    else:
      print(f'Ocorreu um Erro, Opção ínvalida({opc})\n Favor digitar corretamente\n')

# PROGRAMA
main()
```
